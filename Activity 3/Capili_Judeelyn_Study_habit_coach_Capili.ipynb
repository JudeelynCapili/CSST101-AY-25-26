```python
# ============================================
# STUDY HABIT COACH
# Rule-based Expert System for Student Habits
# ============================================

print("📘 WELCOME TO STUDY HABIT COACH 📘")
print("This system will give you advice based on your study habits.\n")

# --- Collect User Inputs ---
sleep_hours = int(input("How many hours do you sleep per day? (e.g., 5, 7, 8): "))
study_hours = int(input("How many hours do you study per day? (e.g., 1, 3, 5): "))
breaks = input("Do you take short breaks while studying? (yes/no): ").strip().lower()
study_time = input("When do you usually study? (morning/afternoon/night): ").strip().lower()
stress_level = input("How is your stress level? (low/medium/high): ").strip().lower()

print("\n🔍 ANALYZING YOUR STUDY HABITS...\n")

# --- Rule 1: Sleep Advice ---
if sleep_hours < 6:
    print("🛌 Rule 1: You sleep less than 6 hours.")
    print("→ Advice: You should aim for at least 7–8 hours of sleep for better focus.")
elif 6 <= sleep_hours <= 8:
    print("🛌 Rule 1: Your sleep schedule is healthy.")
    print("→ Advice: Keep maintaining your consistent sleep routine!")
else:
    print("🛌 Rule 1: You sleep more than 8 hours.")
    print("→ Advice: Try to balance your sleep and study time for productivity.")

# --- Rule 2: Study Hours Advice ---
if study_hours < 2:
    print("\n📖 Rule 2: You study less than 2 hours a day.")
    print("→ Advice: Try to increase your study time gradually for better retention.")
elif 2 <= study_hours <= 4:
    print("\n📖 Rule 2: You study a balanced amount.")
    print("→ Advice: Keep this up and review notes regularly.")
else:
    print("\n📖 Rule 2: You study more than 4 hours a day.")
    print("→ Advice: That’s great, but make sure to rest to avoid burnout.")

# --- Rule 3: Break Habit Advice ---
if breaks == "no":
    print("\n☕ Rule 3: You don’t take breaks while studying.")
    print("→ Advice: Take a 5–10 minute break every hour to improve concentration.")
else:
    print("\n☕ Rule 3: You take regular breaks.")
    print("→ Advice: That’s excellent! Short breaks help your brain reset.")

# --- Rule 4: Study Time Preference ---
if study_time == "morning":
    print("\n🌞 Rule 4: You study in the morning.")
    print("→ Advice: Mornings are great for memory retention — continue this habit.")
elif study_time == "afternoon":
    print("\n🌤 Rule 4: You study in the afternoon.")
    print("→ Advice: Afternoon sessions are productive, but avoid distractions.")
else:
    print("\n🌙 Rule 4: You study at night.")
    print("→ Advice: Nighttime study is fine, but ensure proper rest after studying.")

# --- Rule 5 (CHAINED RULE): Stress Level Based on Habits ---
# This rule depends on both stress level and study habits.
# Example of a chained inference rule.
if stress_level == "high":
    print("\n⚠️ Rule 5: You have a high stress level.")
    if study_hours > 4 or sleep_hours < 6:
        print("→ Chained Advice: Your stress might be caused by overstudying or lack of sleep.")
        print("💡 Try relaxation exercises or reduce study time slightly.")
    else:
        print("→ Chained Advice: Manage stress with light exercise and good nutrition.")
elif stress_level == "medium":
    print("\n😐 Rule 5: You have a moderate stress level.")
    print("→ Advice: Keep balance and take occasional breaks to prevent burnout.")
else:
    print("\n😊 Rule 5: You have low stress levels.")
    print("→ Advice: Excellent! Maintain your routine and stay consistent.")

# --- Final Message ---
print("\n✅ ANALYSIS COMPLETE.")
print("Remember: Consistency, balance, and self-care are keys to effective studying!")
```
